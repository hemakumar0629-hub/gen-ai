```python
import pandas as pd

hikes = pd.Series([10.5, 6.5, 7.5, 5.5],
                   index=['Mani', 'Rahul', 'Jaison', 'Uma'],
                   name='EmployeeHike')
print(hikes)
```

```python
import numpy as np
```

```python
n = 50
df = pd.DataFrame({
    'emp_id' : range(1001, 1001+n),
    'name' : [f'Employee_{i}' for i in range(n)],
    'age' : np.random.randint(25, 40, n),
    'dept' : np.random.choice(['Engineering', 'HR', 'Sales'], n),
    'rating': np.random.choice([1,2,3,4,5], n)
})
df
```

```python
df.head()
```

```python
# Randing choosing the value, Everytime you run the code, the new value will pop up 
np.random.randint(25, 40, n),
```

```python
# Want to take only the rating column 
df['rating'].head()
```

```python
# Want to have two column -> need rating and name column 
df[['name' ,'rating']].head()
```

```python
#Betwwen Function
# Since it is number value - Using betwwen function 
df[df['rating'].between(1,3)]
```
