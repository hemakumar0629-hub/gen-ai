```python
# Queue
# Ticket Booking
# FIFO - First In First Out
# Order Processing 
# Online message -> SMS system
# Avoid Outage -> Slowness -> 8GB (100) 
# In Queue -> 0 1 

from collections import deque
ticket_queue = deque()

ticket_queue.append('Rahul')
ticket_queue.append('Uma')
ticket_queue.append('Jaison')

served = ticket_queue.popleft()
print(served)
print(ticket_queue)
served = ticket_queue.popleft()
print(served)
```
